```python
def get_total_events_massless(r_um, mx_gev, alpha_t, q_thr, exposure):
    outdir = r'/home/yt388/palmer_scratch/data/massless_mediator'
    sphere_type = 'nanosphere' if r_um < 0.5 else 'microsphere'
    
    events = np.empty((mx_gev.size, alpha_t.size))
    events_out = np.empty((mx_gev.size, alpha_t.size))    

    for i, mx in enumerate(mx_gev):
        for j, alpha in enumerate(alpha_t):
            #drdq_file = np.load(outdir + f'/drdq_{sphere_type}_{mx:.5e}_{alpha:.5e}_{m_phi:.0e}.npz', allow_pickle='True')
            drdq_file = np.load(outdir + f'/drdq_{sphere_type}_{R_um:.2e}_{mx:.5e}_{alpha:.5e}.npz', allow_pickle='True')
            
            # GeV; Counts/hour/GeV
            qq, drdq, drdq_out = drdq_file['q'], drdq_file['drdq'], drdq_file['drdq_out']
            events[i][j] = get_nevents(qq, drdq, q_thr, exposure, True)
            events_out[i][j] = get_nevents(qq, drdq_out, q_thr, exposure, True)
    
    return events, events_out
```

```python
def get_nevents(q, drdq, q_thr, exposure, project_z=True):
    # <|cos(theta)|> = 0.5 if q is isotropic
    # so total `q` would look half as small in the z direction
    if project_z:
        q = q * 0.5
        
    drdq[q < q_thr] = 0
    total_rate = np.trapz(drdq, q)
    
    # Exposure in hours
    return total_rate * exposure
```

```python
def get_limit(mx, alpha_n, events):
    cl = 3.0  # 95% CL without background
    
    ret = np.empty_like(mx)
    # Solve for parameters where `events` = cl
    for i, m in enumerate(mx):
        if np.sum(events[i]) < 1e-5:
            ret[i] = np.nan
        else:
            min_alpha_idx = np.argmin( np.abs( np.log10(events[i]) - np.log10((cl)) ))
            
            if events[i][min_alpha_idx] > 1e5:
                ret[i] = np.nan
            else:
                ret[i] = alpha_n[min_alpha_idx]                                  
    return ret
```
