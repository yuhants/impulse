```python
not_nan = np.logical_not(np.isnan(theta))
b = b[not_nan]
theta = theta[not_nan]
```

```python
plt.plot(b, theta, '-')
plt.xscale('log')
```

```python
bcidx = np.argmax(theta)
bcrit = b[bcidx]
```

```python
b1, t1 = b[:bcidx], theta[:bcidx]
b2, t2 = b[bcidx:], theta[bcidx:]

q1 = p * np.sqrt( 2*(1-np.cos(t1)) )
q2 = p * np.sqrt( 2*(1-np.cos(t2)) )
q = p * np.sqrt( 2*(1-np.cos(theta)) )
```

```python
plt.plot(q1, b1, '.-')
plt.plot(q2, b2, '.-')
plt.yscale('log')
```

```python
if(len(b1) > 1 ):
    q1_sorted, b1_sorted, db1 = np.abs( db_dq(q1, b1) )
    
q2_sorted, b2_sorted, db2 = np.abs( db_dq(q2, b2) )
```

```python
plt.plot(q1_sorted, db1)
plt.plot(q2_sorted, db2)
plt.xscale('log')
plt.yscale('log')
```

```python
lnb1 = np.log(b1)
q1_sorted, q1_idx = np.unique(q1, return_index=True)
lnb1_sorted, b1_sorted = lnb1[q1_idx], b1[q1_idx]

lnb2 = np.log(b2)
q2_sorted, q2_idx = np.unique(q2, return_index=True)
lnb2_sorted, b2_sorted = lnb2[q2_idx], b2[q2_idx]
```

So taking `np.unique()` through away some of the events with small `q` that happens when `b` is very large - this is because numerical integration is not accurate enough to resolve differences there. This is fine because these small-q events are much lower than the momentum threshold so will not contribute. 

```python
plt.plot(q1_sorted, b1_sorted)
plt.plot(q2_sorted, b2_sorted)
plt.plot(q1, b1, ':')
plt.plot(q2, b2, ':')
plt.yscale('log')
```

```python
lnb1_cubic = CubicSpline(q1_sorted, lnb1_sorted)
dbdq1 = b1_sorted * lnb1_cubic.derivative()( q1_sorted )

lnb2_cubic = CubicSpline(q2_sorted, lnb2_sorted)
dbdq2 = b2_sorted * lnb2_cubic.derivative()( q2_sorted )
```

Because `b`'s in the first part are so small they barely contribute to the derivatives

```python
plt.plot(q1_sorted, np.abs(dbdq1))
plt.plot(q2_sorted, np.abs(dbdq2))
plt.yscale('log')
plt.xscale('log')
```

```python
plt.plot(q1_sorted, np.abs(2*np.pi*b1_sorted*dbdq1))
plt.plot(q2_sorted, np.abs(2*np.pi*b2_sorted*dbdq2))
plt.xscale('log')
plt.yscale('log')
```
